* __Task 1__: Calculate the product of nonzero elements on the diagonal of a rectangular matrix.  
 For example, for X = np.array([[1, 0, 1], [2, 0, 2], [3, 0, 3], [4, 4, 4]]) answer is 3.

```python
def product_diagonal_python(X):
    rows = len(X)
    cols = len(X[0])

    result = 1
    has_nonzero = False

    for i in range(min(rows, cols)):
        if X[i][i] != 0:
            result *= X[i][i]
            has_nonzero = True

    return result if has_nonzero else 0

def product_diagonal_numpy(X):
    diag = np.diag(X)
    return np.prod(diag[diag != 0]) if np.any(diag != 0) else 0

X = np.array([[1, 0, 1], [2, 0, 2], [3, 0, 3], [4, 4, 4]])
X_list = [[1, 0, 1], [2, 0, 2], [3, 0, 3], [4, 4, 4]]

print(f"np: {product_diagonal_numpy(X)}")
print(f"python: {product_diagonal_python(X_list)}")
```

* __Task 2__: Two vectors x and y are given. Check if they define the same multiset.  
  For example, for x = np.array([1, 2, 2, 4]), y = np.array([4, 2, 1, 2]) answer is True.

```python
from collections import Counter

X_list = [1,2,2,4]
Y_list = [4,2,1,2]
X = np.array([1,2,2,4])
Y = np.array([4,2,1,2])

def same_multiset_python(X,Y):
  return Counter(X) == Counter(Y)

def same_multiset_numpy(x, y):
    vals_x, counts_x = np.unique(x, return_counts=True)
    vals_y, counts_y = np.unique(y, return_counts=True)

    return np.array_equal(vals_x, vals_y) and np.array_equal(counts_x, counts_y)

print(same_multiset_python(X_list, Y_list))
print(same_multiset_numpy(X,Y))
```

* __Task 3__: Find the maximum element in the vector x among the elements preceded by zero.
 For example, for x = np.array([6, 2, 0, 3, 0, 0, 5, 7, 0]) answer is 5.

```python
def max_after_zero_python(X):
    candidates = [X[i + 1] for i in range(len(X) - 1) if X[i] == 0]
    return max(candidates) if candidates else None

def max_after_zero_numpy(X):
    A = X[:-1]
    B = X[1:]

    candidates = B[A == 0]

    return np.max(candidates) if len(candidates) > 0 else None

X = np.array([6,2,0,3,0,0,5,7,0])
X_list = [6,2,0,3,0,0,5,7,0]

print(max_after_zero_numpy(X))
print(max_after_zero_python(X_list))
```

* __Task 4__: Implement run-length encoding. For some vector x, it is necessary to return a tuple of two vectors of the same length. The first one contains numbers and the second one contains how many times they need to be repeated.  
 For example, for x = np.array([2, 2, 2, 3, 3, 3, 5]) answer is (np.array([2, 3, 5]), np.array([3, 3, 1])).

```python
def rle_python(x):
    if len(x) == 0:
        return ([], [])

    values = []
    counts = []

    current_val = x[0]
    count = 1

    for i in range(1, len(x)):
        if x[i] == current_val:
            count += 1
        else:
            values.append(current_val)
            counts.append(count)
            current_val = x[i]
            count = 1

    values.append(current_val)
    counts.append(count)

    return (values, counts)

def rle_numpy(x):
    if len(x) == 0:
        return np.array([]), np.array([])

    changes = np.where(x[1:] != x[:-1])[0] + 1
    idxs = np.concatenate(([0], changes, [len(x)]))

    values = x[idxs[:-1]]
    counts = np.diff(idxs)

    return values, counts

x = np.array([2, 2, 2, 3, 3, 3, 5])
x_list = [2, 2, 2, 3, 3, 3, 5]
print(rle_numpy(x))
print(rle_python(x_list))
```

* __Task 5__: Two samples of objects are given - X and Y. Calculate the matrix of Euclidean distances between objects. Compare with function scipy.spatial.distance.cdist by operating speed.

```python
from scipy.spatial.distance import cdist
import time

def cdist_python(X, Y):
    M = X.shape[0]
    N = Y.shape[0]
    distances = np.zeros((M, N))

    for i in range(M):
        for j in range(N):
            distances[i, j] = np.sqrt(np.sum((X[i] - Y[j]) ** 2))
    return distances

def cdist_numpy(X, Y):
    diff = X[:, np.newaxis, :] - Y
    return np.sqrt(np.sum(diff**2, axis=2))

X = np.random.rand(200, 10)
Y = np.random.rand(300, 10)

t0 = time.time()
res_py = cdist_python(X, Y)
time_py = time.time() - t0

t0 = time.time()
res_np = cdist_numpy(X, Y)
time_np = time.time() - t0

t0 = time.time()
res_scipy = cdist(X, Y, metric="euclidean")
time_scipy = time.time() - t0

print(f"Python: {time_py:.4f} s")
print(f"Numpy: {time_np:.4f} s")
print(f"SciPy: {time_scipy:.5f} s")
```
